```python
import glob
import json
import pandas as pd
import pickle

%matplotlib inline
```

```python
tweets = pd.DataFrame()
```

```python
def process_json(data):
    df = pd.DataFrame(data)
    df = df[(df['lang'] == 'en') | (df['lang'] == 'in') | (df['lang'] == 'zh') | (df['lang'] == 'hi')][['text', 'display_text_range', 'lang']]
    df['text'] = df.apply(lambda row: row['text'][row['display_text_range'][0]:row['display_text_range'][1]] if (type(row['display_text_range']) is list) else row['text'], axis=1)
#     df = df.drop('display_text_range', axis=1)
    return df
```

```python
for s in glob.glob('../bucket/malaysian-tweets/09-16.json'):
    data = json.loads("[%s]" % (open(s,'r').read()[:-2]))
    data = process_json(data)
    tweets = tweets.append(data)
```
